### Functions used in this homework

```python
# https://numpy.org/doc/stable/reference/random/generated/numpy.random.choice.html
# https://www.sharpsightlabs.com/blog/numpy-random-seed/
# https://stackoverflow.com/questions/38138807/how-to-get-the-probability-of-each-vector-belonging-to-each-cluster

def train_test_split_by_class(df, random_state, test_size):
    np.random.seed(random_state)
    B_idx = df.loc[df['Diagnosis']==0].index.values
    M_idx = df.loc[df['Diagnosis']==1].index.values  
    B_test_idx = np.random.choice(B_idx, round(len(B_idx) * test_size), replace=False)
    M_test_idx = np.random.choice(M_idx, round(len(M_idx) * test_size), replace=False)
    test_idx = np.concatenate([B_test_idx, M_test_idx])
    train_idx = np.array([i for i in range(df.shape[0]) if i not in test_idx])
    
    X_train = df.iloc[train_idx, 2:]
    y_train = df.iloc[train_idx, 1]
    X_test = df.iloc[test_idx, 2:]
    y_test = df.iloc[test_idx, 1]
    
    X_train = X_train.astype('float')
    y_train = y_train.astype('int')
    X_test = X_test.astype('float')
    y_test = y_test.astype('int')
    
    return X_train, y_train, X_test, y_test

def label_unlabel_split_by_class(df, random_state, test_size):
    np.random.seed(random_state)
    B_idx = df.loc[df['Diagnosis']==0].index.values
    M_idx = df.loc[df['Diagnosis']==1].index.values  
    B_test_idx = np.random.choice(B_idx, round(len(B_idx) * test_size), replace=False)
    M_test_idx = np.random.choice(M_idx, round(len(M_idx) * test_size), replace=False)
    test_idx = np.concatenate([B_test_idx, M_test_idx])
    train_idx = np.array([i for i in range(df.shape[0]) if i not in test_idx])
    
    X_train = df.iloc[train_idx, 1:]
    y_train = df.iloc[train_idx, 0]
    X_test = df.iloc[test_idx, 1:]
    y_test = df.iloc[test_idx, 0]
    
    X_train = X_train.astype('float')
    y_train = y_train.astype('int')
    X_test = X_test.astype('float')
    y_test = y_test.astype('int')
    
    return X_train, y_train, X_test, y_test

def normalize_data(X_train,X_test):
    transformer = StandardScaler()
    nm_X_train = transformer.fit_transform(X_train)
    nm_X_test = transformer.transform(X_test)
    nm_X_train = pd.DataFrame(data=nm_X_train, columns=X_train.columns)
    nm_X_test = pd.DataFrame(data=nm_X_test, columns=X_test.columns)
    return nm_X_train, nm_X_test

def calculate_metrics(truth, pred, prob):
    report = classification_report(truth, pred, output_dict=True)
    fpr, tpr, thresholds = roc_curve(truth, prob, pos_label=1)
    roc_auc = auc(fpr, tpr)
    return {
        'Accuracy' : report['accuracy'],
        'Precision' : report['weighted avg']['precision'],
        'Recall' : report['weighted avg']['recall'],
        'F1-score' : report['weighted avg']['f1-score'],
        'AUC' : roc_auc
    }

def model_summary(y_true, y_pred, prob):
    print("Confusion Matrix:\n")
    print(confusion_matrix(y_true, y_pred))
    tn, fp, fn, tp = confusion_matrix(y_true,y_pred).ravel()
    
    print("\nThe misclassification in the set:",(fp+fn))
    
    fpr, tpr, thresholds = roc_curve(y_true, prob)
    area = auc(fpr, tpr)
    
    print("Area under the curve:",area)
    print("The accuracy score:",accuracy_score(y_true,y_pred))
    print("The error obtained:",1 - accuracy_score(y_true,y_pred))
    label_="ROC Curve : area = {} ".format(round(area, 4))
    fig, axes = plt.subplots()
    axes.plot(fpr, tpr, label=label_)
    axes.plot([0,1],[0,1])
    axes.set_xlabel("False positive rate")
    axes.set_ylabel("True positive rate")
    axes.set_title("ROC AUC Curve")
    axes.legend()
    plt.show()
    
def semi_supervised_model(X_labeled, y_labeled, X_unlabeled, clf):
    X_unlabeled = np.array(X_unlabeled)
    while(len(X_unlabeled)>0):
        dist = np.absolute(clf.decision_function(X_unlabeled))
        farthest_idx = np.argmax(dist)
        farthest_pt = np.array(X_unlabeled[farthest_idx]).reshape(1,30)
        farthest_pt_predict = clf.predict(farthest_pt)
        X_unlabeled = np.delete(X_unlabeled, farthest_idx, axis=0)
        X_labeled = np.append(X_labeled, farthest_pt, axis=0)
        y_labeled = np.append(y_labeled, farthest_pt_predict, axis=0)
        clf.fit(X_labeled, y_labeled)
    return clf

def get_pred_proba(centers, X_data, cluster_label,y_data):
    X_data = np.array(X_data)
    y_data = np.array(y_data)
    label_cluster = np.ndarray((2,), dtype='int')
    majority_label = cluster_label.copy()
    proba = np.ndarray((len(X_data), 2))
    
    for cluster in [0, 1]:
        sample_idx, = np.where(cluster_label == cluster)
        top_30_idx = np.argsort(cdist(X_data, centers, 'euclidean')[:,cluster], axis=0)[:30]
        top_30_labels = y_data[top_30_idx]
        majority = Counter(top_30_labels).most_common(1)[0][0]
        majority_label[sample_idx] = majority
        label_cluster[majority] = cluster
    
    for i, x in enumerate(X_data):
        proba[i, 0] = np.linalg.norm(x - centers[0])
        proba[i, 1] = np.linalg.norm(x - centers[1])
    proba = softmax(proba, axis=1).astype(np.float64)
    
    pos_cluster_ = label_cluster[0]
    proba = proba[:, pos_cluster_]
    
    return majority_label, proba

def get_pred_proba_spectral(X_data, cluster_label,y_data):
    X_data = np.array(X_data)
    y_data = np.array(y_data)
    label_cluster = np.ndarray((2,), dtype='int')
    majority_label = cluster_label.copy()
    proba = np.ndarray((len(X_data), 2))
    
    for cluster in [0, 1]:
        sample_idx, = np.where(cluster_label == cluster)
        all_labels = y_data[sample_idx]
        majority = Counter(all_labels).most_common(1)[0][0]
        majority_label[sample_idx] = majority
        label_cluster[majority] = cluster
    
    dist_from_0 = np.mean(cdist(X_data, X_data[cluster_label==0], 'euclidean'),axis=1)
    dist_from_1 = np.mean(cdist(X_data, X_data[cluster_label==1], 'euclidean'),axis=1)

    for i, x in enumerate(X_data):
        proba[i, 0] = dist_from_0[i]
        proba[i, 1] = dist_from_1[i]
    
    proba = softmax(proba, axis=1).astype(np.float64)
    pos_cluster_ = label_cluster[0]
    proba = proba[:, pos_cluster_]
    
    return majority_label, proba
```

### i. Supervised Learning: Train an L1-penalized SVM to classify the data. Use 5 fold cross validation to choose the penalty parameter. Use normalized data. Report the average accuracy, precision, recall, F1-score, and AUC, for both training and test sets over your M runs. Plot the ROC and report the confusion matrix for training and testing in one of the runs.

```python
# https://scikit-learn.org/stable/modules/generated/sklearn.svm.LinearSVC.html

grid = {'C': np.logspace(-4, 5, 10)}
split = StratifiedKFold(5, random_state=42, shuffle=True)
train_summary = []
test_summary = []
for i in range(1,31):
    
    # split data
    X_train, y_train, X_test, y_test = train_test_split_by_class(df=df,random_state=i,test_size=0.2)
    
    # normalize data
    X_train, X_test = normalize_data(X_train,X_test)
    
    # evaluate model
    
    # training data
    linear_svc = LinearSVC(penalty='l1',dual=False)
    grid_cv = GridSearchCV(linear_svc, grid, cv=split, n_jobs=-1)
    grid_cv.fit(X_train, y_train)
    
    print('\033[1m'+"\nIteration:",i,'\033[0m')
    print("\nBest C:",grid_cv.best_params_['C'])
    best_grid_cv = grid_cv.best_estimator_
    
    y_pred_train = best_grid_cv.predict(X_train)
    y_train_prob = best_grid_cv.decision_function(X_train)
    
    #test data
    y_pred_test = best_grid_cv.predict(X_test)
    y_test_prob = best_grid_cv.decision_function(X_test)
    
    train_metric = calculate_metrics(y_train,y_pred_train,y_train_prob)
    print("Train accuracy:",train_metric['Accuracy'])
    train_summary.append(train_metric)
    test_metric = calculate_metrics(y_test,y_pred_test,y_test_prob)
    print("Test accuracy:",test_metric['Accuracy'])
    test_summary.append(test_metric) 
train_summary_df = pd.DataFrame(train_summary)
test_summary_df = pd.DataFrame(test_summary)
train_avg_df = train_summary_df.apply(lambda x: x.mean(), axis=0)
test_avg_df = test_summary_df.apply(lambda x: x.mean(), axis=0)
supervised_avg_summary = pd.concat([train_avg_df,test_avg_df],axis=1,keys=['Train Average','Test Average']).T
```

### iii. Unsupervised Learning: Run k-means algorithm on the whole training set. Ignore the labels of the data, and assume k = 2.

### Part A, B, C are all answered by the following code.

#### Answer A:

We can avoid being trapped in local minimum by setting a stopping criteria i.e. the number of iterations in kmeans. Here, I have set "n_init=15" to avoid the kmeans algorithm to get trapped in local minimum.

```python
# https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html

train_summary = []
test_summary = []
for i in range(1,31):
    X_train, y_train, X_test, y_test = train_test_split_by_class(df=df,random_state=i,test_size=0.2)
    X_train, X_test = normalize_data(X_train,X_test)
    
    kmeans = KMeans(n_clusters=2,random_state=i,n_init = 15,n_jobs=-1)
    train_cluster_labels = kmeans.fit_predict(X_train)
    
    print('\033[1m'+"\nIteration:",i,'\033[0m')
    
    # for train data
    
    y_pred_train, y_train_prob = get_pred_proba(kmeans.cluster_centers_, X_train, train_cluster_labels,y_train)
    train_metric = calculate_metrics(y_train,y_pred_train,y_train_prob)
    print("Train accuracy:",train_metric['Accuracy'])
    train_summary.append(train_metric)
    
    # for test data
    
    test_cluster_labels = kmeans.predict(X_test)
    y_pred_test, y_test_prob = get_pred_proba(kmeans.cluster_centers_, X_test, test_cluster_labels,y_test)  
    test_metric = calculate_metrics(y_test,y_pred_test,y_test_prob)
    print("Test accuracy:",test_metric['Accuracy'])
    test_summary.append(test_metric) 
      
train_summary_df = pd.DataFrame(train_summary)
test_summary_df = pd.DataFrame(test_summary)
train_avg_df = train_summary_df.apply(lambda x: x.mean(), axis=0)
test_avg_df = test_summary_df.apply(lambda x: x.mean(), axis=0)
unsupervised_avg_summary = pd.concat([train_avg_df,test_avg_df],axis=1,keys=['Train Average','Test Average']).T
```

#### Summary

```python
print("Average Summary:\n")
print(unsupervised_avg_summary)
print('\n')
print("="*100)
print("\nROC and Confusion matrix of the last (30th) Monte Carlo run:\n")
print("Training set:\n")
model_summary(y_train,y_pred_train,y_train_prob)
print("Test set:\n")
model_summary(y_test,y_pred_test,y_test_prob)
```

### iv. Spectral Clustering: Repeat 1(b)iii using spectral clustering, which is clustering based on kernels. Research what spectral clustering is. Use RBF kernel with gamma=1 or find a gamma for which the two clutsres have the same balance as the one in original data set (if the positive class has p and the negative class has n samples, the two clusters must have p and n members). Do not label data based on their proximity to cluster center, because spectral clustering may give you non-convex clusters . Instead, use fit - predict method.

For gamma=1, I was getting imbalanced classes, hence I tried different gamma values and chose 1e-3 which gave balanced classes. Also, I have used knn with number of neighbors=1 for test prediction as Spectral clustering do not offer ".predict".

```python
# https://github.com/scikit-learn/scikit-learn/issues/901
# https://scikit-learn.org/stable/modules/generated/sklearn.cluster.SpectralClustering.html
# https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html

train_summary = []
test_summary = []
for i in range(1,31):
    X_train, y_train, X_test, y_test = train_test_split_by_class(df=df,random_state=i,test_size=0.2)
    X_train, X_test = normalize_data(X_train,X_test)
    
    model = SpectralClustering(n_clusters=2, random_state=i,n_init = 15,n_jobs=-1,affinity='rbf', gamma=1e-3)
    train_cluster_labels = model.fit_predict(X_train)
    
    print('\033[1m'+"\nIteration:",i,'\033[0m')
    
#     for train data
    
    y_pred_train, y_train_prob = get_pred_proba_spectral(X_train, train_cluster_labels,y_train)
    train_metric = calculate_metrics(y_train,y_pred_train,y_train_prob)
    print("Train accuracy:",train_metric['Accuracy'])
    train_summary.append(train_metric)
    
#     for test data

    knn = KNeighborsClassifier(n_neighbors=1)
    knn.fit(X_train, y_pred_train)
    y_pred_test = knn.predict(X_test)
    y_test_prob = knn.predict_proba(X_test)
    test_metric = calculate_metrics(y_test,y_pred_test,y_test_prob[:,1])
    print("Test accuracy:",test_metric['Accuracy'])
    test_summary.append(test_metric) 
      
train_summary_df = pd.DataFrame(train_summary)
test_summary_df = pd.DataFrame(test_summary)
train_avg_df = train_summary_df.apply(lambda x: x.mean(), axis=0)
test_avg_df = test_summary_df.apply(lambda x: x.mean(), axis=0)
spectral_avg_summary = pd.concat([train_avg_df,test_avg_df],axis=1,keys=['Train Average','Test Average']).T
```

#### Summary

```python
print("Average Summary:\n")
print(spectral_avg_summary)
print('\n')
print("="*100)
print("\nROC and Confusion matrix of the last (30th) Monte Carlo run:\n")
print("Training set:\n")
model_summary(y_train,y_pred_train,y_train_prob)
print("Test set:\n")
model_summary(y_test,y_pred_test,y_test_prob[:,1])
```

### (b) Repeat each of the following two procedures 50 times. You will have 50 errors for 90 SVMs per each procedure.

### i. Train a SVM with a pool of 10 randomly selected data points from the training set using linear kernel and L1 penalty. Select the penalty parameter using 5-fold cross validation. Repeat this process by adding 10 other randomly selected data points to the pool, until you use all the 900 points. Do NOT replace the samples back into the training set at each step. Calculate the test error for each SVM. You will have 90 SVMs that were trained using 10, 20, 30, ... , 900 data points and their 90 test errors. You have implemented passive learning.

```python
passive_accuracy = []
passive_error = []
grid = {'C' : np.logspace(-5, 4, 20)}
for i in range(1,51):
    X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, test_size=0.344, random_state=i)
    X_train, X_test = normalize_data(X_train,X_test)
    
    split = StratifiedKFold(5, random_state=i, shuffle=True)
    
    unused_idx = np.arange(len(X_train))
    used_idx = []
    metrics = {'accuracy': [], 'error': []}
    
    for __ in range(90):
        pool_idx = random.sample(unused_idx.tolist(), 10)
        while len(Counter(y_train.iloc[pool_idx]).keys())==1:
            pool_idx = random.sample(unused_idx.tolist(), 10)
        unused_idx = np.setdiff1d(unused_idx, np.array(pool_idx))
        used_idx.extend(pool_idx)

        X_train_pool = X_train.iloc[used_idx] 
        y_train_pool = y_train.iloc[used_idx]

        linear_svc = LinearSVC(penalty='l1',dual=False)
        grid_cv = GridSearchCV(linear_svc, grid, cv=split, n_jobs=-1)
        grid_cv.fit(X_train_pool, y_train_pool)

        best_grid_cv = grid_cv.best_estimator_

        y_pred = best_grid_cv.predict(X_test)
        test_summary = classification_report(y_test, y_pred, output_dict=True)
        
        metrics['accuracy'].append(test_summary['accuracy'])
        metrics['error'].append(1-test_summary['accuracy'])
        
    print('\033[1m'+"\nIteration:",i,'\033[0m')    
    print("Test error with 10 data points:",metrics['error'][0])
    print("Test error with 900 data points:",metrics['error'][len(metrics['error'])-1])
        
    passive_accuracy.append(metrics['accuracy'])
    passive_error.append(metrics['error'])
```

### ii. Train a SVM with a pool of 10 randomly selected data points from the training set using linear kernel and L1 penalty. Select the parameters of the SVM with 5-fold cross validation. Choose the 10 closest data points in the training set to the hyperplane of the SVM and add them to the pool. Do not replace the samples back into the training set. Train a new SVM using the pool. Repeat this process until all training data is used. You will have 90 SVMs that were trained using 10, 20, 30,..., 900 data points and their 90 test errors. You have implemented active learning.

```python
active_accuracy = []
active_error = []
grid = {'C' : np.logspace(-5, 4, 20)}
for i in range(1,51):
    X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, test_size=0.344, random_state=i)
    X_train, X_test = normalize_data(X_train,X_test)
    
    split = StratifiedKFold(5, random_state=i, shuffle=True)
    
    unused_idx = np.arange(len(X_train))
    used_idx = []
    metrics = {'accuracy': [], 'error': []}
    
    pool_idx = random.sample(unused_idx.tolist(), 10)
    while len(Counter(y_train.iloc[pool_idx]).keys())==1:
        pool_idx = random.sample(unused_idx.tolist(), 10)
    
    for __ in range(90):
        unused_idx = np.setdiff1d(unused_idx, np.array(pool_idx))
        used_idx.extend(pool_idx)
        
        X_train_pool = X_train.iloc[used_idx] 
        y_train_pool = y_train.iloc[used_idx]
        
        linear_svc = LinearSVC(penalty='l1',dual=False)
        grid_cv = GridSearchCV(linear_svc, grid, cv=split, n_jobs=-1)
        grid_cv.fit(X_train_pool, y_train_pool)

        best_grid_cv = grid_cv.best_estimator_

        y_pred = best_grid_cv.predict(X_test)
        test_summary = classification_report(y_test, y_pred, output_dict=True)

        metrics['accuracy'].append(test_summary['accuracy'])
        metrics['error'].append(1-test_summary['accuracy'])
        
        # taking closest 10 data points 
        distances = best_grid_cv.decision_function(X_train.iloc[unused_idx])
        pool_idx = np.argsort(abs(distances))[:10]
    
    print('\033[1m'+"\nIteration:",i,'\033[0m')    
    print("Test error with 10 data points:",metrics['error'][0])
    print("Test error with 900 data points:",metrics['error'][len(metrics['error'])-1])
        
    active_accuracy.append(metrics['accuracy'])
    active_error.append(metrics['error'])
```

### (c) Average the 50 test errors for each of the incrementally trained 90 SVMs in 2(b)i and 2(b)ii. By doing so, you are performing a Monte Carlo simulation. Plot average test error versus number of training instances for both active and passive learners on the same figure and report your conclusions. Here, you are actually obtaining a learning curve by Monte-Carlo simulation.

```python
x = np.arange(10, 901, 10)
fig, axes = plt.subplots(figsize=(12, 7))
axes.plot(x, np.mean(passive_error,axis=0), label='Passive')
axes.plot(x, np.mean(active_error,axis=0), label='Active')
axes.set_xlabel("Size of Training set")
axes.set_ylabel("Average Test Error")
axes.legend()
axes.set_title("Monte Carlo Learning Curve")
plt.show()
```

**Conclusions:**



- For passive learning, the average test error initially drops rapidly, then gradually decreases as the number of training sample increases.
- For active learning, the average test error initially drops rapidly more than passive learning curve, then gradually decreases as the number of training sample increases.
- Thus we can say that, initially active learning model performs better than passive learning model as the test error drops faster, but as we go on adding more training samples, both the learning models performs in similar way.
